```python
import pandas as pd
import seaborn as sns

path = r'C:\Users\Supervisor031\DE_Sprint\Введение\DataEngineer.csv'
df = pd.read_csv(path)
# print(df.head(60))

vacancies_20 = df[['Company Name', 'Job Title']].groupby(['Company Name']).count().reset_index().sort_values('Job Title', ascending=False)[:20]
vacancies_20['Company Name'] = vacancies_20['Company Name'].apply(lambda x: x.split('\n')[0])
print(vacancies_20)
now = sns.barplot(data=vacancies_20, y='Job Title', x='Company Name', palette="crest")
now.tick_params(axis='x', rotation=90)
now.set(xlabel='Company Name', ylabel='Amount')
now.set_title('Первые 20 компаний с самым большим кол-вом вакансий.')
```

```python
names_vacancies = df[['Job Title', 'Salary Estimate']].groupby(['Job Title']).count().reset_index().sort_values('Salary Estimate', ascending=False)[:20]
print(names_vacancies)
now = sns.barplot(data=names_vacancies, y='Salary Estimate', x='Job Title', palette="crest")
now.tick_params(axis='x', rotation=90)
now.set(xlabel='Job Title', ylabel='Amount')
now.set_title('Первые 20 самых популярных названий в вакансиях.')
```

```python
names_owner = df[['Type of ownership', 'Salary Estimate']].groupby(['Type of ownership']).count().reset_index().sort_values('Salary Estimate', ascending=False)[:20]
print(names_owner)
now = sns.barplot(data=names_owner, y='Salary Estimate', x='Type of ownership', palette="crest")
now.tick_params(axis='x', rotation=90)
now.set(xlabel='Type of ownership', ylabel='Amount')
now.set_title('Первые 20 самых популярных Type of ownership, относительно численности вакансий.')
```

```python
names_industry = df[['Industry', 'Salary Estimate']].groupby(['Industry']).count().reset_index().sort_values('Salary Estimate', ascending=False)[:20]
print(names_industry)
now = sns.barplot(data=names_industry, y='Salary Estimate', x='Industry', palette="crest")
now.tick_params(axis='x', rotation=90)
now.set(xlabel='Industry', ylabel='Amount')
now.set_title('Первые 20 самых популярных разделов по секторам работы.')
```

```python
def avg_salary(temp):
    if len(temp) == 2:
        x = temp[0].strip('$K')
        y = temp[1][:temp[1].find('K')].strip('$')
        return round((int(x) + int(y)) / 2, 1)
    else:
        return temp


df['temp_salary'] = df['Salary Estimate'].apply(lambda x: x.split('-'))
df['avg_salary'] = df['temp_salary'].apply(avg_salary)

names_industry = df[['Industry', 'avg_salary']].groupby(['Industry']).mean().reset_index().sort_values('avg_salary', ascending=False)[:20]
print(names_industry)
now = sns.barplot(data=names_industry, y='avg_salary', x='Industry', palette="crest")
now.tick_params(axis='x', rotation=90)
now.set(xlabel='Industry', ylabel='Amount')
now.set_title('Первые 20 популярных секторов работы и средний уровень зарплат в них.')
```

```python
temp = df[['Industry', 'Rating']].groupby(['Industry']).mean().reset_index().sort_values('Rating', ascending=False)[:20]
print(temp)

now = sns.barplot(data=temp, y='Rating', x='Industry', palette="crest")
now.tick_params(axis='x', rotation=90)
now.set(xlabel='Industry', ylabel='Amount')
now.set_title('Гистограмма распределения рейтинга компаний по отраслям.')
```

```python
temp = df.loc[:, ['Company Name', 'avg_salary']]
temp['new_name'] = temp['Company Name'].apply(lambda x: x.split('\n')[0])

new_df = temp[['new_name', 'avg_salary']].groupby(['new_name']).count().reset_index().sort_values('avg_salary', ascending=False)
new_df.rename(columns={"avg_salary": "count_vac"}, inplace=True)

temp = temp.merge(new_df, on='new_name', how='left')

final = temp[temp['count_vac'] >= 10].groupby('new_name').mean().reset_index().sort_values('avg_salary', ascending=False)
print(final)

now = sns.barplot(data=final, y='avg_salary', x='new_name', palette="crest")
now.tick_params(axis='x', rotation=90)
now.set(xlabel='new_name', ylabel='Amount')
now.set_title('График средних зарплат по популярным компаниям с количеством вакансий >= 10.')
```

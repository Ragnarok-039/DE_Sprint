```python
import pandas as pd
import json


def salary(my_s):
    my_s = [i.strip() for i in str(my_s).split('-')]
    while 'None' in my_s:
        my_s.remove('None')
    if len(my_s) == 0:
        return 0
    elif len(my_s) == 1:
        return int(my_s[0])
    else:
        if int(my_s[0]) >= int(my_s[1]):
            return int(my_s[0])
        else:
            return int(my_s[1])


with open('vacancies.json') as file:
    data = json.load(file)

df = pd.json_normalize(data, record_path='data')

df['max_salary'] = df['salary'].apply(salary)

print(df.describe())
print(df.head())
print('-' * 30)

print('Средняя зарплата по каждому региону.')
print(round(df.groupby('region').mean(), 1))
print()

print('Медианная зарплата по каждому региону.')
print(round(df.groupby('region').median(), 1))
print()

print('Самая высокооплачиваемая из групп вакансий, исходя из их средних зарплат:')
print(*list(df[df['max_salary'] >= df['max_salary'].mean()]['title']), sep='\n')
print()

print('Процентное соотношение каждого региона по вакансиям от всех вакансий:')
print(df.groupby('region').count()['title'] / df['title'].count() * 100)
```

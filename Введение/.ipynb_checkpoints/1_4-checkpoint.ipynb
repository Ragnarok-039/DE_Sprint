```python
# 1

import numpy as np
import matplotlib.pyplot as plt

data = np.random.rayleigh(scale=2, size=(1, 50000))

plt.hist(data[0], bins=50)
plt.show()
```

```python
# 2

import sympy as sp
import numpy as np

sp.var('x y')

now = sp.Eq((x ** 2 + y ** 2 - 1) ** 3 - (x ** 2) * (y ** 3), 0)

pli = sp.plot_implicit(now, show=False)
series = pli[0]
data, action = series.get_points()
data = np.array([(x_int.mid, y_int.mid) for x_int, y_int in data])

plt.scatter(data[:, 0], data[:, 1])
plt.grid()
plt.show()
```

```python
# 3

import numpy as np
import matplotlib.pyplot as plt

age = []
sex = []
weight = []
children = []
smoke = []
located = []
price = []

with open('dataset_home.txt') as file:
    for now in file:
        now = now.split(',')
        age.append(int(now[0].strip()))
        sex.append(now[1].strip())
        weight.append(float(now[2].strip()))
        children.append(int(now[3].strip()))
        smoke.append(now[4].strip())
        located.append(now[5].strip())
        price.append(float(now[-1].strip()))
#         print(now)

age = np.array(age)
sex = np.array(sex)
weight = np.array(weight)
children = np.array(children)
smoke = np.array(smoke)
located = np.array(located)
price = np.array(price)

# print(price)
# print(weight)

plt.hist(price, bins=50)
plt.xlabel('Цена страховки')
plt.ylabel('Количество')
plt.grid(True)
plt.show()

plt.scatter(weight, price)
plt.xlabel('ИМТ')
plt.ylabel('Цена страховки')
plt.grid(True)
plt.show()

plt.scatter(age, price)
plt.xlabel('Возраст')
plt.ylabel('Цена страховки')
plt.grid(True)
plt.show()

plt.bar(children, price)
plt.xlabel('Количество детей')
plt.ylabel('Цена страховки')
plt.show()

plt.bar(smoke, price)
plt.xlabel('Курение')
plt.ylabel('Цена страховки')
plt.show()

plt.bar(located, price)
plt.xlabel('Месторасположение')
plt.ylabel('Цена страховки')
plt.show()

plt.bar(sex, price)
plt.xlabel('Пол')
plt.ylabel('Цена страховки')
plt.show()
```

```python
# 1

import numpy as np
import matplotlib.pyplot as plt

data = np.random.rayleigh(scale=2, size=(1, 50000))

plt.hist(data[0], bins=50)
plt.show()
```

```python
# 2

import sympy as sp
import numpy as np

sp.var('x y')

now = sp.Eq((x ** 2 + y ** 2 - 1) ** 3 - (x ** 2) * (y ** 3), 0)

pli = sp.plot_implicit(now, show=False)
series = pli[0]
data, action = series.get_points()
data = np.array([(x_int.mid, y_int.mid) for x_int, y_int in data])

plt.scatter(data[:, 0], data[:, 1])
plt.grid()
plt.show()
```

```python
# 3

import numpy as np
import matplotlib.pyplot as plt

price = []
weight = []

with open('dataset_home.txt') as file:
    for now in file:
        now = now.split(',')
        price.append(float(now[-1].strip()))
        weight.append(float(now[2]))
#         print(now)

price = np.array(price)
weight = np.array(weight)

print(price)
print(weight)

plt.hist(price)
plt.show()

plt.scatter(price, weight)
plt.show()
```

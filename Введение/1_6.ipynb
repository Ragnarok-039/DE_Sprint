```python
import pandas as pd
import seaborn as sns

path = r'C:\Users\Supervisor031\DE_Sprint\Введение\DataEngineer.csv'
df = pd.read_csv(path)
# print(df.head(60))

vacancies_20 = df[['Company Name', 'Job Title']].groupby(['Company Name']).count().reset_index().sort_values('Job Title', ascending=False)[:20]
vacancies_20['Company Name'] = vacancies_20['Company Name'].apply(lambda x: x.split('\n')[0])
print(vacancies_20)
now = sns.barplot(data=vacancies_20, y='Job Title', x='Company Name', palette="crest")
now.tick_params(axis='x', rotation=90)
now.set(xlabel='Company Name', ylabel='Amount')
now.set_title('Первые 20 компаний с самым большим кол-вом вакансий.')
```

```python
def avg_salary(temp):
    if len(temp) == 2:
        x = temp[0].strip('$K')
        y = temp[1][:temp[1].find('K')].strip('$')
        return round((int(x) + int(y)) / 2, 1)
    else:
        return temp


df['temp_salary'] = df['Salary Estimate'].apply(lambda x: x.split('-'))
df['avg_salary'] = df['temp_salary'].apply(avg_salary)

names_vacancies = df[['Industry', 'avg_salary']].groupby(['Industry']).mean().reset_index().sort_values('avg_salary', ascending=False)[:20]
print(names_vacancies)
now = sns.barplot(data=names_vacancies, y='avg_salary', x='Industry', palette="crest")
now.tick_params(axis='x', rotation=90)
now.set(xlabel='Industry', ylabel='Amount')
now.set_title('Первые 20 секторов работы и средний уровень зарплат в них.')
```
